```python
def fill_missing_sector(row):
    if pd.isnull(row['sector']):
        if 'financial' in row['sector_description'].lower() or 'bank' in row['sector_description'].lower() or 'fintech' in row['sector_description'].lower():
            return 'Finance'
        elif 'medical' in row['sector_description'].lower() or 'health' in row['sector_description'].lower():
            return 'Health'
        elif 'vehicle' in row['sector_description'].lower() or 'transport' in row['sector_description'].lower():
            return 'Transport'
        elif 'games' in row['sector_description'].lower() or 'sports' in row['sector_description'].lower():
            return 'Game'
        elif 'brand marketing' in row['sector_description'].lower() or 'marketing' in row['sector_description'].lower():
            return 'Marketing'
        elif 'farming' in row['sector_description'].lower() or 'fishery' in row['sector_description'].lower():
            return 'Agriculture'
        elif 'real eastate' in row['sector_description'].lower() or 'home design' in row['sector_description'].lower() or 'accommodation' in row['sector_description'].lower():
            return 'Real Estate'
        elif 'energy' in row['sector_description'].lower() or 'battery' in row['sector_description'].lower() or 'solar' in row['sector_description'].lower():
            return 'Energy'
        elif 'school' in row['sector_description'].lower() or'education' in row['sector_description'].lower():
            return 'Education'
        elif 'fashion' in row['sector_description'].lower() or 'd2c fashion' in row['sector_description'].lower():
            return 'Style'
        elif 'entertainment' in row['sector_description'].lower() or 'media' in row['sector_description'].lower() or 'news' in row['sector_description'].lower():
            return 'Media'
        else:
            return 'Other'
    return row['sector']
```

```python
import re
def sector_arrangement(sector):
    if re.search(r'bank|fintech|finance|crypto|account|credit|venture|crowd|blockchain', sector, re.IGNORECASE):
        return 'Finance'
    elif re.search(r'automotive|air transport|transport|logistics|vehicle|transportation', sector, re.IGNORECASE):
        return 'Transport'
    elif re.search(r'health|medical|pharmaceutical|health diagnostics|health care|hospital|biotechnology|biopharma|hygiene management|healthtech|laboratory|biotech', sector, re.IGNORECASE):
        return 'Health'
    elif re.search(r'brand marketing|marketing|event promotion|market', sector, re.IGNORECASE):
        return 'Marketing'
    elif re.search(r'farming|agriculture|nutrition|hauz khas|fishery', sector, re.IGNORECASE):
        return 'Agriculture'
    elif re.search(r'e-learning|edtech|education|communities|preschool|school', sector, re.IGNORECASE):
        return 'Education'
    elif re.search(r'real estate|home decor|accommodation|interior design|home design|home interior|co-living|construction', sector, re.IGNORECASE):
        return 'Real Estate'
    elif re.search(r'renewable energy|solar|energy|energy storage|battery|oil and gas|renewable|ev|electricity|purification|renewables', sector, re.IGNORECASE):
        return 'Energy'
    elif re.search(r'entertainment|media|news|video|digital media|fm|publication|ott|social network|music|digital platform|podcast|publishing|blogging', sector, re.IGNORECASE):
        return 'Media'
    elif re.search(r'fashion|jewelry|eyeglasses|eye wear|apparel|dating|celebrity|eyewear|d2c fashion|matrimony|textiles|d2c jewellery|footwear|arts', sector, re.IGNORECASE):
        return 'Style'
    elif re.search(r'gaming|games|sports', sector, re.IGNORECASE):
        return 'Game'
    else:
        return sector
```

```python
df['Sector'] = df['Sector'].apply(sector_arrangement)
```

```python
#  Cleaning the HeadQuarter column
# Within the HeadQuarter column lies some info that do not belong, hence we remve them and replace with null values
df['HeadQuarter'] = df['HeadQuarter'].str.replace('Pharmaceuticals\t#REF!','', regex= False)
df['HeadQuarter'] = df['HeadQuarter'].str.replace('Information Technology & Services','', regex= False)
df['HeadQuarter'] = df['HeadQuarter'].str.replace('Online Media\t#REF!','', regex= False)
df['HeadQuarter'] = df['HeadQuarter'].str.replace('Computer Games','', regex= False)
df['HeadQuarter'] = df['HeadQuarter'].str.replace('Food & Beverages','', regex= False)
df['HeadQuarter'] = df['HeadQuarter'].str.replace('Gurugram\t#REF!','', regex= False)


df['HeadQuarter'] = df['HeadQuarter'].fillna(df['HeadQuarter'].mode(), axis= 0)
df['HeadQuarter'] = df['HeadQuarter'].apply(lambda x: x.split(',')[0])
```
